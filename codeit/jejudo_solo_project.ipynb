### df_2017 과 df_2018을 합치기
- pandas 에서 수직('axis=0')이나 수평('axis=1')으로 연결할 수 있다.  
> #### df1 과 df2 를 수직으로 합치는 코드  
가정: df1과 df2는 같은 열 이름을 가지고 있습니다.  
> *import pandas as pd  
result_df = pd.concat([df1, df2], axis=0)*
- 또한 ignore_index=True 매개변수를 추가하면 합친 결과 데이터프레임의 인덱스가 새로운 순차적 정수 인덱스로 재설정됩니다.

# 카페 이용자수와 방문자수(유동인구)의 관계 알아보기
> 1. 카페 이용자수를 알아보기 위해 '비알콜 음료점업'들만 있는 df를 추출한다.
> 2. 방문자수가 포함되어 있는 df_population 에서 '시군구명'과 '성별'로 그룹화하여 추출한다.
> 3. 두 그래프를 index를 동일하게 만들어 주었으니 cafe_by_area에 '방문인구'라는 새로운 col을 생성후, 
population_by_area의 '방문인구' col의 값을 for문을 통해 cafe_by_area에 합쳐준다.
> 4. corr(numeric_only=True)를 통해 상관계수를 확인하고, sns.heatmap()을 통해 시각화 시킨다.

```python
#카페가 속한 업종명을 찾기위해 값들을 확인한다
df_all['업종명'].unique()
```

```python
#위 과정에서 찾은 '비알콜 음료점업'을 통해 새로운 df 추출
df_cafe = df_all[df_all['업종명'] == '비알콜 음료점업']
df_cafe
```

```python
#위에서 만든 df_cafe를 sum 해주어 정리
cafe_by_area = df_cafe.groupby(['시군구명', '성별']).sum(numeric_only=True).reset_index()
cafe_by_area
```

```python
#위와 동일한 index와 순서를 맞춰주기위해 groupby 후 sum해주어 정리
population_by_area = df_population.groupby(['시군구명', '성별']).sum(numeric_only=True).reset_index()
population_by_area
```

##### 실수를 한 부분. 방문인구의 값들을 넘겨주는 것이 아니라 그냥 merge를 했기 때문에 index가 늘어나고 corr()를 사용할 수 없는 df의 형태가 나옴.
> 이 실수를 통해 '방문인구'의 값들을 넘겨주는 방법을 찾음

```python
merged_df = pd.merge(cafe_by_area, population_by_area, on='시군구명')
correlation = merged_df['이용자수'].corr(merged_df['방문인구'])
correlation
```

### cafe_by_area 라는 df에 '방문인구'라는 새로운 col을 생성해준 후, for문을 사용하여 값 추가해주기

```python
cafe_by_area['방문인구'] = pd.Series(dtype='float64')
for index in cafe_by_area.index:
    cafe_by_area.loc[index, '방문인구'] = population_by_area.loc[index, '방문인구']
    
cafe_by_area
```

### df.corr()을 통해서 상관계수를 보고 sns.heatmap을 사용해 시각화 한다.

```python
cafe_by_area.corr(numeric_only=True)
```

```python
sns.heatmap(cafe_by_area.corr(numeric_only=True))
```
