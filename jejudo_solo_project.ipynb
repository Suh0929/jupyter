### df_2017 과 df_2018을 합치기
- pandas 에서 수직('axis=0')이나 수평('axis=1')으로 연결할 수 있다.  
> #### df1 과 df2 를 수직으로 합치는 코드  
가정: df1과 df2는 같은 열 이름을 가지고 있습니다.  
> *import pandas as pd  
result_df = pd.concat([df1, df2], axis=0)*
- 또한 ignore_index=True 매개변수를 추가하면 합친 결과 데이터프레임의 인덱스가 새로운 순차적 정수 인덱스로 재설정됩니다.  
> *result_df = pd.concat([df1, df2], axis=0, ignore_index=True)*
- 만약 두 데이터프레임에 중복된 데이터가 있거나 중복을 피하고 싶다면, drop_duplicates() 함수를 호출하여 중복된 행을 제거할 수 있습니다.
> *result_df = pd.concat([df1, df2], axis=0, ignore_index=True).drop_duplicates()*

```python
df_all = pd.concat([df_2017, df_2018], axis=0, ignore_index=True)
df_all
```
