> * 객체형 데이터나 카테고리형 데이터에 대해서는 유니크한 값의 개수, 가장 자주 등장하는 값(top), 해당 값이 데이터에서 등장하는 빈도(frequency) 등을 보여줍니다.

###  columns 들중 dtype이 'object' 인 것들만 뽑아내 인덱스 객체형태로 정리한다.
> * df.dtypes는 DataFrame의 각 컬럼의 데이터 타입을 반환하며, df.dtypes == 'object'는 
이 데이터 타입이 'object'인지 여부를 불리언(Boolean) 값으로 반환합니다. 

> * 이 불리언 배열을 사용하여 df.columns[] 인덱싱을 통해 'object' 데이터 타입을 가진 모든 컬럼의 이름을 선택한다.

```python
object_columns = df.columns[df.dtypes == 'object']
object_columns
```

### for문을 사용하여 object_columns 리스트 안에 col들과 각 col에 포함된 값들을 정리한다.
> print(df[col].unique(), \n)를 통해서 각 col의 고유값들을 뽑아내어 정리한다.

```python
for col in object_columns:
    print(col)
    print(df[col].unique(), '\n')
```

### '연월' col을 '-'을 기준으로 split 하여 첫번째 값을 df['연도']에 새롭게 저장한다.
> * split() 메소드는 ()안에 들어간 값을 기준으로 분리하며, 리스트 형태로 반환한다. 또한 str 타입에만 사용가능하다.

```python
df['연도'] = df['연월'].str.split('-').str[0]
df['연도']
```

### '연도' col의 값들을 문자열에서 int dtype으로 변경해준다
* astype() 은 ()의 타입을 입력하면 그 타입으로 변경해주는 메소드이다.

```python
df['연도'] = df['연도'].astype(int)
```

### '연도' col 에서 2016년도는 제거해주는 과정

```python
df = df[df['연도'] != 2016]
```

```python
df.shape
```

```python
for col in object_columns:
    print(col)
    print(df[col].unique(), '\n')
```

### 연월 단위로 그룹화 시켜주고, sum 해준후 인덱스를 초기화 시켜주기

```python
groupby_ym = df.groupby('연월').sum(numeric_only=True).reset_index()
groupby_ym
```

```python
sns.barplot(data=groupby_ym, x='연월', y='이용금액')
plt.rc('font', family='NanumGothic')
plt.rcParams['figure.figsize'] = (10, 5)
plt.xticks(rotation=90)
plt.title('연월별 카드 이용 금액')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/100000000:,.0f}'))
```

### barplot으로 위에 데이터를 그래프화 시키고 가독성을 높이는 과정
* *폰트 설정과 크기설정 - plt.rc('font', )*
> plt.rc는 "run command"의 약어로, Matplotlib의 구성(configuration) 옵션을 동적으로 설정할 수 있게 해주는 함수입니다. 이를 통해 폰트, 선 스타일, 색상 등과 같은 그래프의 전역 설정을 컨트롤 할 수 있습니다

* *x축 제목을 90도 돌려서 겹치지 않게 만들기*
> plt.xticks 함수는 X축의 눈금 위치와 레이블을 설정할 때 사용됩니다. 이 함수는 눈금의 위치와 해당 위치에 표시될 레이블을 지정할 수 있으며, 추가적으로 레이블의 회전 각도나 폰트 속성 등을 조정할 수 있는 인자도 제공합니다. 예를 들어, plt.xticks(rotation=90)은 X축의 모든 눈금 레이블을 90도 회전시켜 세로로 표시하게 합니다.

* *title을 달아서 보기 쉽게 해준다*
>  plt.title('')  
* *y축의 단위를 가독성이 좋게 1억 단위로 나누고 소숫점자리를 지정해 시각화 한다.*
> plt.gca는 "get current axes"의 약어로, 현재 활성화된 축(axes) 객체를 반환합니다. 이 축 객체를 통해 특정 축의 속성을 직접 수정할 수 있습니다. 예를 들어, plt.gca().yaxis.set_major_formatter()와 같이 사용하여 y축의 주요 눈금 포맷을 변경할 수 있습니다.

* *그래프의 크기를 조절해 가독성을 높인다.*
> plt.rcParams는 Matplotlib에서 전역 그래프 설정을 관리하는 구성 딕셔너리(dictionary)입니다.  이 딕셔너리를 통해 Matplotlib의 다양한 스타일 및 동작 옵션을 설정할 수 있으며, 이는 모든 그래프에 영향을 미치게 됩니다.  plt.rcParams를 사용하면 선의 두께, 폰트 속성, 그래프의 DPI, 그리드 존재 유무, 색상 스키마 등 많은 시각적 요소를 사전에 정의하고, 이를 일관되게 적용할 수 있습니다. 이러한 방식은 특정 프로젝트 전체에서 일관된 시각적 스타일을 유지하고자 할 때 매우 유용합니다.
>> eg) plt.rcParams['lines.linewidth'] = 2  
plt.rcParams['font.size'] = 14  
plt.rcParams['figure.figsize'] = [10, 5]

### 연령대 별 카드 이용자 수 비중을 구하기 시작

```python
groupby_age = df.groupby('연령대').sum(numeric_only=True).reset_index()
groupby_age
```

```python
groupby_age.plot(kind='pie', y='이용자수', labels=groupby_age['연령대'])
plt.title('연령대별 카드 이용자 수 비중')
```

### codeit페이지에서는 20대미만이 0번 index에 해당되어 있어 바꾸는 방법을 찾아봄
> * 임시 변수를 사용하여 행을 스왑하는 방법  
>> 1. groupby_age.loc[0] = groupby_age.loc[1]  
groupby_age.loc[1] = temp  
>> 2. 인덱스 리스트를 가져옵니다.  
index_list = groupby_age.index.tolist()  
0번 인덱스와 1번 인덱스를 스왑합니다.  
index_list[0], index_list[1] = index_list[1], index_list[0]  
데이터프레임의 인덱스를 새로운 순서로 업데이트합니다.  
groupby_age = groupby_age.reindex(index_list)


```python
# 임시 변수를 사용해 행을 스왑합니다.
temp = groupby_age.loc[0].copy()
groupby_age.loc[0] = groupby_age.loc[1]
groupby_age.loc[1] = temp
groupby_age
```

### 크기가 비슷한 데이터들이 있어 숫자로 표시하기 위해 autopct 메소드 사용

```python
groupby_age.plot(kind='pie', y='이용자수', labels=groupby_age['연령대'], autopct='%.1f%%')
plt.title('연령대별 카드 이용자 수 비중')
```

### 위에 그래프는 범례가  그래프와 겹쳐져 있어 가독성이 떨어지므로 범례의 위치를 수정
> - plt.legend() 를 사용시 일반적으로 사용할 수 있는 위치 옵션:   
'upper right', 'upper left', 'lower left', 'lower right', 'right', 'center left', 'center right', 'lower center', 'upper center', 'center'.  
> - 내가 헷갈린 부분    
*bbox_to_anchor로 설정한 위치는 범례의 기준점을 정하는 박스의 위치로 생각할 수 있다.  
그리고 loc 매개변수는 이 박스 내에서 범례의 상대적인 위치를 조정한다다.  
즉, bbox_to_anchor가 범례를 위치시킬 전체 그래프 대비 좌표를 설정하고, loc는 이 좌표 내에서 범례가 어떻게 정렬될지를 정하는 것이다.*

```python
groupby_age.plot(kind='pie', y='이용자수', labels=groupby_age['연령대'], autopct='%.1f%%')
plt.title('연령대별 카드 이용자 수 비중')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

### 인당이용금액을 구하기 위해서 새로운 col을 생성하고 그 col에 이용금액 / 이용자수를 해준다.

```python
groupby_age['인당이용금액'] = groupby_age['이용금액'] / groupby_age['이용자수']
groupby_age
```

### 20대 미만의 금액은 너무 작아 억단위로 표현하면 가독성이 떨어지므로 단위는 유지하면서 ,만 추가해줌

### for문을 통해서 연령대별 (업종별 인당이용 금액)을 pie 그래프를 그리기

```python
for age in groupby_age_market['연령대'].unique():
    data = groupby_age_market[groupby_age_market['연령대'] == age]
    data.plot(y='인당이용금액', labels=data['업종명'], kind='pie', autopct='%.1f%%')
    plt.title(f'업종별 인당 이용 금액({age})')
    plt.xticks(rotation=90)
    plt.legend(loc ='best')
    plt.show()
```

### 연령대별 연월에 따른 이용금액을 barplot 으로 나타내기

```python
groupby_age_ym = df.groupby(['연령대', '연월']).sum(numeric_only=True).reset_index()
sns.barplot(data=groupby_age_ym, x = '연령대', y= '이용금액', hue='연월')
```

### 가독성이 매우 떨어지므로 위에 for문을 사용한 것처럼 연령대별 barplot으로 변경

```python
for age in groupby_age_ym['연령대'].unique():
    data = groupby_age_ym[groupby_age_ym['연령대'] == age]
    sns.barplot(data=data, x='연월', y='이용금액')
    plt.title(f'{age} 연월별 카드 이용 금액')
    plt.xticks(rotation=90)
    plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x/100000000:,.0f}'))
    plt.ylabel('이용금액(억)')
    plt.show()
```
