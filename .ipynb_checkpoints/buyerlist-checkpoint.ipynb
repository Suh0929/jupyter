```python
pvc_flooring = ['PVC FLOORING', 'PVC FLOOIRNG', 'PVC FLOORING\nCD-ROM', 'PVC FLOORING\nCATALOGUE', 'PVC SPONGE FLOORING\nPVC FLOORING\nCD-ROM', 
                'PVC SHEET', 'COMPACT DISK', 'PVC FLOOIRNG\nMOSQUITO ZAPPER', 'PVC FLOORING\nCD-ROM\nCOIN MAT', 'CATALOGUE', 
                'PVC FLOORING\nCOMPACT DISK', 'SAMPLE BOOK', 'CD-ROM', 'FLOORING SAMPLE\nCD-ROM', 'FLOORING SAMPLE', 'PVC SHEET SAMPLE', 
                'SPONGE FLOORING SAMPLE', 'PVC FLOORING\nPVC FLOORING PATTERN CD', 'PVC FLOORING\nTILE SAMPLE BOOK', 'PVC FLOORING \nCATALOGUE', 
                'PVC FILM', 'PVC FOAM SHEET', 'PVC FLOORING \nCATALOGUE\nCD-ROM']
```

```python
def check_contains(sample):
    for pvc_string in pvc_flooring:
        if pvc_string in sample:
            return 'PVC FLOORING'
    return sample

df['발송했던 샘플'] = df['발송했던 샘플'].apply(check_contains)
df
```
